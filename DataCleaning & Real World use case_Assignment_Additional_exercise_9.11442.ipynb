# EXPLORING DATA USING SEABORN

```python
from sklearn import datasets
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
iris = datasets.load_iris()
```

```python
iris
```

```python
iris.data
```

```python
iris.target
```

```python
iris.feature_names
```

```python
data = pd.DataFrame(data=iris.data,columns=iris.feature_names)
```

```python
iris.target
```

```python
iris.target_names
```

```python
label = pd.DataFrame(list(map(lambda x : iris.target_names[x],iris.target)),columns=['Species'])
```

```python
iris = pd.concat([data,label],axis=1)
```

```python
iris.head()
```

```python
iris.rename(index=str, columns={"sepal length (cm)": 'sepallength',"sepal width (cm)": "sepalwidth","petal length (cm)":"petallength","petal width (cm)":"petalwidth","Species":"species"},inplace=True)
```

```python
iris.head()
```

#### 1) Execute the displot() to see the distribution of the SepalLength,Sepalwidth,petallength and petalwidth features. 
#### Plot them as a subplot in a single image

```python
#using plt.Subplots 
fig,axes = plt.subplots(2,2,figsize=(6,6))
sns.distplot(iris[('sepallength')],ax=axes[0,0],kde=False)
sns.distplot(iris['sepalwidth'],ax=axes[0,1],kde=False)
sns.distplot(iris['petallength'],ax=axes[1,0],kde=False)
sns.distplot(iris['petalwidth'],ax=axes[1,1],kde=False)
```

```python
#using plt.Subplot
sns.distplot(iris[('sepallength')],ax=plt.subplot(2,2,1),kde=False)
sns.distplot(iris[('sepalwidth')],ax=plt.subplot(2,2,2),kde=False)
sns.distplot(iris[('petallength')],ax=plt.subplot(2,2,3),kde=False)
sns.distplot(iris[('petalwidth')],ax=plt.subplot(2,2,4),kde=False)
```

#### 3) Do a count plot for Feature Species

```python
sns.countplot(x='species',data=iris)
```

#### 4) Do a pairplot on the features 'sepallength','sepalwidth','petallength','petalwidth' and 'species'

```python
sns.set_style('darkgrid')
sns.pairplot(iris,hue='species',size=4)
```

#### 5) Do an lmplot on the following sepallength,petallength. using hue, display the different species in the different colors.

```python
sns.lmplot('sepallength','petallength',iris,hue='species')
```

#### 6) Do a barplot of "species" vs "sepallength".

```python
sns.barplot(x='species',y='sepallength',data=iris)
```

#### 7) Using heatmap, plot the correlation matrix.

```python
corr = iris.corr()
sns.heatmap(corr)
```
