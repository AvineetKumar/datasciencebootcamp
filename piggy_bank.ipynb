```python
Bank=0

def add_money(x):
    global Bank
    Bank=Bank+x
    print("Amount Added",x)
    
def withdraw(x):
    global Bank
    Bank=Bank-x
    print("amount withdrawn",x)
    
def checkbalance():
    global Bank
    balance=Bank
    print("your balance is",balance)
    
def menu(a):
    if a==1:
        b=float(input("Enter amount"))
        if b>=1:
            add_money(b)
        else:
            print("enter positive amount")
        select()
    elif a==2:
        c=float(input("Pleaase enter amount: "))
        if c>=1:
            withdraw(c)
        else:
            print("Enter positive amount : ")
        select()
    elif a==3:
        checkbalance()
        select()
    
    elif a==4:
        print("Good Byeee")
        piggybank()
    else:
        print("please enter correct option")
        select()
        
def select():
    l=input('''select your option from below\n 1-Add Money\n 2-Withdraw Money \n 3-Check Balance\n 4-Exit\n''')
    
    try:
        l=float(l)
        menu(l)
    except ValueError:
        print("Please enter correct option")
        select()
        
def piggybank():
    
    start= input("Do you want to Start or Stop\n")
    if start.lower() in 'start':
        print("Welcome to Piggy Bank")
        select()
    elif start.lower() in 'stop':
        print("Good Bye")
        #exit
    else:
        print("Enter the correct option")
        piggybank()
```
