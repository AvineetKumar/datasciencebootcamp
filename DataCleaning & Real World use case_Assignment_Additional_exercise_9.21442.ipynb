# Matplotlib Excercises

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

#### Exercise1

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
```

#### Exercise2

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.2,0.2])
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.2,0.2])

ax1.plot(x,y,'b')
ax1.set_xlabel("X")
ax1.set_ylabel("Y")

ax2.plot(x,y,'r')
ax2.set_xlabel("x")
ax2.set_ylabel("y")
```

#### Exercise 3

```python
abc = plt.figure()

abc.add_axes([0,0,1,1])
abc.add_axes([0.2,0.5,0.4,0.4])
```

```python
abc = plt.figure()

ax2 = abc.add_axes([0,0,1,1])
ax3 = abc.add_axes([0.2,0.5,0.4,0.4])

ax2.plot(x,z)
ax2.set_xlabel('X')
ax2.set_ylabel('Z')

ax3.plot(x,y)
ax3.set_xlabel('X')
ax3.set_ylabel('Y')
```

#### Exercise4

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y,'b',linewidth=3.00,ls='--')
axes[1].plot(x,z,'r',linewidth=3.00)
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(10,2))
axes[0].plot(x,y,'b',linewidth=3.00,ls='--')
axes[1].plot(x,z,'r',linewidth=3.00)
```
